```python
Final_df.dropna(subset=['Abstract', 'Authors','Year'], inplace=True)
Final_df.isnull().sum()
```

```python
# Convert the Year column from float to integer
Final_df['Year'] = Final_df['Year'].astype(int)
Final_df.dtypes
```

```python
# Count of duplicate rows
duplicate_count = Final_df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
```

```python
# Remove duplicate rows across all columns
Final_df = Final_df.drop_duplicates()

# Reset index after dropping duplicates, if desired
#df.reset_index(drop=True, inplace=True)

# Check the shape of the DataFrame to confirm duplicates were removed
print("DataFrame shape after removing duplicates:", Final_df.shape)
```

```python
# Count of duplicate rows
duplicate_count = Final_df.duplicated().sum()
print(f"Number of duplicate rows after removing: {duplicate_count}")
```

```python
# # Count duplicate rows based on the 'Title' and 'Authors' columns
duplicate_count = Final_df.duplicated(subset=['Title', 'Authors', 'Abstract']).sum()
print(f"Total duplicate rows based on Title, Authors & Abstract is: {duplicate_count}")
```

```python
# Remove duplicate rows based on the 'Title' and 'Authors' columns
Final_df = Final_df.drop_duplicates(subset=['Title', 'Authors','Abstract'], keep='first')

# Verify if duplicates are removed
print(f"DataFrame shape after removing duplicates: {Final_df.shape}")
```

```python
Final_df = Final_df.drop(['DOI', 'URL','PMID'], axis=1)
Final_df.head()
```
